Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
doc1 = "The weather in Rotkreuz is nice. Rotkreuz is big."
doc2 = "The campus is in Rotkreuz"
doc3= "The Department of Informatics at Hochschule Luzern"
```

```python
documents = [doc1,doc2,doc3]
```

```python
vectorizer = CountVectorizer()
```

```python
print(vectorizer.fit_transform(documents))
```

```python
print(vectorizer.fit_transform(documents).todense())

print(vectorizer.vocabulary_)
```

Cosine Similarity

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

doc1 = "I like movies with a lot of action."
doc2 = "I like movies with a lot of action."
doc3 = " I hate movies with a romantic touch."

documents = [doc1, doc2, doc3]
```

```python
count_vectorizer = CountVectorizer(stop_words='english')
count_vectorizer = CountVectorizer()
sparse_matrix = count_vectorizer.fit_transform(documents)
```

```python
print(sparse_matrix.todense())
print(count_vectorizer.get_feature_names_out())
```

```python
count_vectorizer_all = CountVectorizer(token_pattern=r"(?u)\b\w+\b")
sparse_matrix_all = count_vectorizer_all.fit_transform(documents)

print(count_vectorizer_all.get_feature_names_out())
```

```python
count_vectorizer_stop = CountVectorizer(stop_words='english')
sparse_matrix_stop = count_vectorizer_stop.fit_transform(documents)

print(count_vectorizer_stop.get_feature_names_out())
```

```python
df = pd.DataFrame(
    sparse_matrix_stop.todense(), 
    columns=count_vectorizer_stop.get_feature_names_out(), 
    index=['doc1', 'doc2', 'doc3']
)
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

print(cosine_similarity(df, df))
```

TF-IDF with Sklearn 

```python
from sklearn.feature_extraction.text import TfidfVectorizer 

doc1 = "I like movies."
doc2 = "Movies are fun."
doc3 = "Movies are cool."
```

```python
documents = [doc1, doc2, doc3]
```

```python
tfidf = TfidfVectorizer()
result = tfidf.fit_transform(documents)
```

```python
print(result)
```

```python
print('\nWord indexes:')
print(tfidf.vocabulary_)
```

```python
tfidf.get_feature_names_out()
```

```python
tfidf.vocabulary_
```

```python
print('\nidf values:')
for ele1, ele2 in zip(tfidf.get_feature_names_out(), tfidf.idf_):
    print(ele1, ':', ele2)
```

```python
print(result.toarray())
```

```python
print('\ntf-idf values in matrix form:')
print(result.toarray())
```

```python
df = pd.DataFrame(
    result.toarray(), 
    columns=tfidf.get_feature_names_out(), 
    index=['doc1', 'doc2', 'doc3']
)
```
