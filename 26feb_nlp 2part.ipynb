```python
import nltk
from nltk.tokenize import RegexpTokenizer, WordPunctTokenizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

text = "When Alexander Graham Bell invented the telephone, he had three missed calls from Chuck Norris."

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
```

```python
tokenizer1 = RegexpTokenizer('\S+')
tokens_whitespace = tokenizer1.tokenize(text)
print(tokens_whitespace)

tokenizer2 = RegexpTokenizer('[A-Z]\w+')
tokens_capital = tokenizer2.tokenize(text)
print(tokens_capital)
```

```python
stop_words = set(stopwords.words('english'))
tokenizer_std = WordPunctTokenizer()
all_tokens = tokenizer_std.tokenize(text)

filtered_text = [w for w in all_tokens if not w.lower() in stop_words]
print(filtered_text)
```

```python
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("Stemming:")
for w in all_tokens:
    print(w, ":", ps.stem(w))

print("\nLemmatization:")
for w in all_tokens:
    print(w, ":", lemmatizer.lemmatize(w, pos="v"))
```
