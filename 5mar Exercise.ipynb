```python
datei = open('IMDB_1.txt','r')
imdb1 = datei.read()
datei.close()
print(imdb1)
datei = open('IMDB_2.txt','r')
imdb2 = datei.read()
datei.close()
print(imdb2)
datei = open('IMDB_3.txt','r')
imdb3 = datei.read()
datei.close()
print(imdb3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import WordPunctTokenizer
```

1. Build a Bag of Words model, which does not include stop words from the documents IMDB_1, IMDB_2 and IMDB_3.

```python
imdb = [imdb1, imdb2, imdb3]
```

```python
vectorizer = CountVectorizer(stop_words='english')
bow = vectorizer.fit_transform(imdb)
```

```python
print(vectorizer.get_feature_names_out())
```

2. Compute the TF-IDF score using the documents IMDB_1, IMDB_2 and IMDB_3 for the two terms: Rick and artifacts. 

```python
tfidf_vec = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vec.fit_transform(imdb)
```

```python
feature_names = tfidf_vec.get_feature_names_out()
rick_idx = list(feature_names).index('rick')
arti_idx = list(feature_names).index('artifacts')
```

```python
for i, doc in enumerate(imdb):
    print(f"Doc {i+1} - Rick: {tfidf_matrix[i, rick_idx]:.4f}")
    print(f"Doc {i+1} - Artifacts: {tfidf_matrix[i, arti_idx]:.4f}")
```

with stopwords

```python
tfidf_vec_with_stop = TfidfVectorizer()
tfidf_matrix_with_stop = tfidf_vec_with_stop.fit_transform(imdb)
```

```python
feature_names_stop = tfidf_vec_with_stop.get_feature_names_out()
```

```python
rick_idx_s = list(feature_names_stop).index('rick')
arti_idx_s = list(feature_names_stop).index('artifacts')
```

```python
for i, doc in enumerate(imdb):
    print(f"Doc {i+1} - Rick: {tfidf_matrix_with_stop[i, rick_idx_s]:.4f}")
    print(f"Doc {i+1} - Artifacts: {tfidf_matrix_with_stop[i, arti_idx_s]:.4f}")
```
