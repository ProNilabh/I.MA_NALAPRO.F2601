```python
import nltk
from nltk.tokenize import WordPunctTokenizer
from nltk.tokenize import NLTKWordTokenizer
from nltk.tokenize import TreebankWordTokenizer
```

```python
sentence = "it's raining outside"
tokenizer1 = WordPunctTokenizer()
tokenizer1.tokenize (sentence)
```

```python
sentence = "it's raining outside"
tokenizer2 = NLTKWordTokenizer()
tokenizer2.tokenize (sentence)
```

```python
sentence = "it's raining outside"
tokenizer3 = TreebankWordTokenizer()
tokenizer3.tokenize (sentence)
```
