## 1. N-Grams
Identify the Bi- and Trigrams from 'shakespeare-macbeth.txt’. This book is part of the Gutenberg Corpus. It can be loaded directly into NLTK. The dataset must be cleaned first (Stop Words, Punctuation, etc.)

```python
import nltk
import string
from nltk.corpus import gutenberg, stopwords
from nltk.util import ngrams
from collections import Counter
```

```python
nltk.download('gutenberg')
nltk.download('stopwords')
```

```python
macbeth_words = gutenberg.words('shakespeare-macbeth.txt')
```

```python
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
```

```python
cleaned_words = [
    word.lower() for word in macbeth_words 
    if word.lower() not in stop_words 
    and word not in punctuation 
    and word.isalpha()
]
```

```python
bigrams = list(ngrams(cleaned_words, 2))
trigrams = list(ngrams(cleaned_words, 3))
```

```python
print("Top 5 Bigrams:", Counter(bigrams).most_common(5))
```

```python
print("ALL BIGRAMS")
for gram in bigrams:
    print(gram)

print("\nALL TRIGRAMS")
for gram in trigrams:
    print(gram)
```

### 2. TF-IDF

Compute the TF-IDF value for Bigrams from the documents IMDB_1.txt, IMDB_2.txt, IMDB_3.txt

```python
import pandas as pd
import string
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
file_names = ['IMDB_1.txt', 'IMDB_2.txt', 'IMDB_3.txt']
```

```python

# 2. Text Cleaning Function
def clean_text(text):
    # Remove punctuation to prevent it from splitting bigrams incorrectly
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Lowercase the text for uniformity
    return text.lower()

# 3. Load and Preprocess the Corpus
corpus = []
for file_name in file_names:
    with open(file_name, 'r') as f:
        content = f.read()
        corpus.append(clean_text(content))

# 4. Compute TF-IDF for Bigrams
# ngram_range=(2, 2) specifically targets Bigrams
# stop_words='english' removes common words like 'the', 'is', 'and'
vectorizer = TfidfVectorizer(ngram_range=(2, 2), stop_words='english', use_idf=True)
tfidf_matrix = vectorizer.fit_transform(corpus)
feature_names = vectorizer.get_feature_names_out()

# 5. Create a DataFrame for the results
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names, index=file_names)

# Display the Transposed DataFrame (Bigrams as rows)
print(tfidf_df.T)

# Optional: Save to CSV for business reporting
tfidf_df.T.to_csv('bigram_tfidf_results.csv')
```
