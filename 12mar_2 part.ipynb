```python
from nltk import word_tokenize
from gensim.models import Word2Vec

text = """World of Warcraft (WoW) is a massively multiplayer online role-playing game.
Released in 2004 by Blizzard Entertainment.
Set in the Warcraft fantasy universe World of Warcraft takes place within the world of Azeroth.
approximately four years after the events of the previous game
The game was announced in 2001, and was released for the 10th anniversary of the Warcraft franchise on November 23, 2004.
Since launch, World of Warcraft has had eight major expansion packs: The Burning Crusade (2007), Wrath of the Lich King (2008), Cataclysm (2010), Mists of Pandaria (2012),
Warlords of Draenor (2014), Legion (2016), Battle for Azeroth (2018), and Shadowlands (2020). """

tokens = [word_tokenize(s) for s in text.lower().split('\n')]

print(tokens)
```

```python
model = Word2Vec(tokens, min_count=1, window=2, vector_size=40, workers=2, epochs=5)
print(model)
```
